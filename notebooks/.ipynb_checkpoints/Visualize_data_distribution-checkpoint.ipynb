```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.special import comb
```

```python
data_path = "../Data/Data_prepared/"
```

```python
data_list = pd.read_csv("/blue/juannanzhou/ProteinLLE/Data/Data_prepared/Protein_set.txt", header=None)
data_list = [file.split(".")[0] for file in data_list.iloc[:, 0]]
```

```python
# get conversion dictionary from old full data name to shorthand

data_summary = pd.read_csv("../results/Dataset summary-Table 1.csv")

name_dict = {}
for i in range(len(data_summary)):
    name_dict[data_summary.iloc[i]['OldName']] = data_summary.iloc[i]['Name'].split("_")[1]
```

```python
data_list
```

```python
# pick a dataset

data_name = data_list[0]
print(data_name)
datafile = pd.read_csv(data_path + data_name + ".csv")
```

```python
if 'Faure' in data_name: 
    alpha = 2
else: alpha = 20
# how many sites are variable? 

seqs = datafile.mutated_sequence
seqs = [list(seq) for seq in seqs]
seqs = pd.DataFrame(seqs)

summary = {}
for i in range(seqs.shape[1]):
    summary[i] = seqs[i].unique()

lens = []
for i in range(seqs.shape[1]):
    lens.append(len(summary[i]))
    
L = np.sum(np.array(lens) != 1)
print(f"alpha = {alpha}")
print(f"number of variable sites = {L}")


pos_var = np.where(np.array(lens) != 1)[0]
```

```python
n_data_dist_log = []

for d in range(L + 1):
    count = np.sum(datafile.hd == d)
    # if count == 0:
    #     count += 10**-10
    if d == 0:
        count = 1
    n_data_dist_log.append(np.log10(count))
n_data_dist_log = np.array(n_data_dist_log)
    
n_seqs_dist_log = [np.log10(comb(L, d)*(alpha - 1)**d) for d in range(0, L+1)] # number of sequences per distance class
n_seqs_dist_log = np.array(n_seqs_dist_log)
props_dist_log = n_data_dist_log - n_seqs_dist_log # log (no. of datapoints/size of distance class)
```

```python
# prepare plot data

x_max = min(L, 35)

xticks = np.arange(0, x_max, 5)

hist_data = datafile.hd[datafile.hd < x_max]

log_prop_data = props_dist_log[:x_max + 1]

y_lo = -1*round(-1*np.min(log_prop_data[log_prop_data != -np.inf]))

y_lo = np.min(- n_seqs_dist_log[:x_max + 1])

yticks = [1, 10**-2, 10**-4, 10**-6, 10**-8, 10**-10, 10**-12, 10**-14, 10**-16]
```

```python
plot_data = {}

for data_name in data_list:
    
    if data_name == 'Somermeyer2022_allGFP':
        pass
    else:
        
        print(data_name)
        datafile = pd.read_csv(data_path + data_name + ".csv")

        if 'Faure' in data_name: 
            alpha = 2
        else: alpha = 20
        # how many sites are variable? 

        seqs = datafile.mutated_sequence
        seqs = [list(seq) for seq in seqs]
        seqs = pd.DataFrame(seqs)

        summary = {}
        for i in range(seqs.shape[1]):
            summary[i] = seqs[i].unique()

        lens = []
        for i in range(seqs.shape[1]):
            lens.append(len(summary[i]))

        L = np.sum(np.array(lens) != 1)
        x_max = min(L, 35)

        pos_var = np.where(np.array(lens) != 1)[0] 

        n_data_dist_log = []
        for d in range(x_max + 1):
            count = np.sum(datafile.hd == d)
            # if count == 0:
            #     count += 10**-10
            if d == 0:
                count = 1
            n_data_dist_log.append(np.log10(count))
        n_data_dist_log = np.array(n_data_dist_log)

        n_seqs_dist_log = [np.log10(comb(L, d)*(alpha - 1)**d) for d in range(0, x_max+1)] # number of sequences per distance class
        n_seqs_dist_log = np.array(n_seqs_dist_log)
        props_dist_log = n_data_dist_log - n_seqs_dist_log # log (no. of datapoints/size of distance class)    

        plot_data[data_name] = {"n_seqs_dist_log":n_seqs_dist_log,  "props_dist_log": props_dist_log, "x_max": x_max, 
                               "hist_data": datafile.hd[datafile.hd < x_max], "log_prop_data": props_dist_log[:x_max + 1], 
                               "y_lo": np.min(-1*n_seqs_dist_log[n_seqs_dist_log != -np.inf])}
```

```python
if y_lo < 20:
    yticks = [1] + [10**-5, 10**-10, 10**-15, 10**-20]
    
else: yticks = [1] + [10**-10, 10**-20, 10**-35, 10**-40, 10**-60, 10**-80, 10**-100]
```

```python
# Create subplots with shared x-axis
fig, axs = plt.subplots(2, 5, sharex='col', gridspec_kw={'hspace': 0.15, 'wspace': 0.4})

for i in range(5):
    data_name = list(plot_data.keys())[i]
    hist_data = plot_data[data_name]['hist_data']
    log_prop_data = plot_data[data_name]['log_prop_data']
    n_seqs_dist_log  = plot_data[data_name]['n_seqs_dist_log']
    y_lo = plot_data[data_name]['y_lo']
    x_max = plot_data[data_name]['x_max']
    if x_max > 30:
        xticks = np.arange(0, x_max + 1, 10)
    else: xticks = np.arange(0, x_max + 1, 5)
    
    
    if -1*y_lo < 5:
        yticks = [1] + [10**-1, 10**-2, 10**-3, 10**-4, 10**-5]    
    elif -1*y_lo < 20:
        yticks = [1] + [10**-5, 10**-10, 10**-15, 10**-20, 10**-25, 10**-30, 10**-35, 10**-40]    
    elif -1*y_lo > 60:
        yticks = [1] + [10**-20, 10**-40, 10**-60, 10**-80, 10**-100]
        
    else: yticks = [1] + [10**-10, 10**-20, 10**-30, 10**-40, 10**-50, 10**-60, 10**-70]
        
    yticks = yticks[:np.where(y_lo - 1 < np.log10(np.array(yticks)))[0][-1] + 1]
    axs[0, i].hist(hist_data, bins = 20, color='grey')
    axs[0, i].set_title(name_dict[data_name])
    
    # Plot line plot on the second subplot
    axs[1, i].scatter(range(len(log_prop_data)), 10**log_prop_data, color=(0.25, 0.25, 0.25))
    axs[1, i].plot(10**(-n_seqs_dist_log[:x_max + 1]), color='gray')
    axs[1, i].set_yscale('log')

    # Customize y-tick labels to show as "10^n"
    axs[1, i].tick_params(axis='y', which='minor', labelsize=0)  # Hide minor tick labels
    axs[1, i].set_yticks(yticks)
    # axs[1, i].set_ylim(10**y_lo, 10**0)
    axs[1, i].set_xticks(xticks)
    axs[1, i].set_xlabel('Distance')
    # Set common x-axis label
    if i%5 == 0:

        axs[0, i].set_ylabel('Count')
        axs[1, i].set_ylabel('Proportion')


fig.set_size_inches(6.5*5, 5*2)  # Adjust the size as needed

plt.savefig('../figures/' + 'data_dist_a.pdf')
# Show the plot
plt.show()

```
