```python
import scipy.stats as stats

# Create a contingency table
# Replace the values below with your actual data
# |                  | Group 1   | Group 2   |
# |------------------|-----------|-----------|
# | Subjects with    | x1        | x2        |
# | the property     |           |           |
# | Subjects without | n1 - x1   | n2 - x2   |
# | the property     |           |           |
x1 = 3  # Number of subjects with the property in Group 1
x2 = 5  # Number of subjects with the property in Group 2
n1 = 20  # Total number of subjects in Group 1
n2 = 14  # Total number of subjects in Group 2

# Perform Fisher's exact test
odds_ratio, p_value = stats.fisher_exact([[x1, n1 - x1], [x2, n2 - x2]])

# Print the results
print(f"Odds Ratio: {odds_ratio}")
print(f"P-value: {p_value}")
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.multicomp import MultiComparison

# Define your data for the control group and treatment groups
# Replace these with your actual data
control_group = np.array([20, 25, 30, 35])
treatment_group_1 = np.array([25, 28, 32, 36])  # Treatment group 1
treatment_group_2 = np.array([22, 27, 31, 37])  # Treatment group 2
# Add more treatment groups as needed

# Combine all data and create corresponding group labels
data = np.concatenate([control_group, treatment_group_1, treatment_group_2])
groups = ['Control'] * len(control_group) + ['Treatment 1'] * len(treatment_group_1) + ['Treatment 2'] * len(treatment_group_2)

# Perform Dunnett's test
mc = MultiComparison(data, groups)
result = mc.tukeyhsd()
print(result)
```

```python
import numpy as np
control = np.array([7.40, 8.50, 7.20, 8.24, 9.84, 8.32])
drug_a = np.array([9.76, 8.80, 7.68, 9.36])
drug_b = np.array([12.80, 9.68, 12.16, 9.20, 10.55])

import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
ax.boxplot([control, drug_a, drug_b])
ax.set_xticklabels(["Control", "Drug A", "Drug B"])  
ax.set_ylabel("mean")  
plt.show()

from scipy.stats import dunnett
res = dunnett(drug_a, drug_b, control=control)
res.pvalue
```
