### Evidence

```python
def binom(n, k):
    """Compute binomial coefficient using the log-gamma function for stability."""
    # Ensure n and k are tensors
    n = torch.as_tensor(n, dtype=torch.float)
    k = torch.as_tensor(k, dtype=torch.float)
    return torch.exp(torch.lgamma(n + 1) - torch.lgamma(k + 1) - torch.lgamma(n - k + 1))
```

```python
y = preds[0]

y = (y - y.mean())/y.std()

import random
import gpytorch
sub_list = random.sample(range(len(datafile)), 20000)
train_list = random.sample(sub_list, int(.5*len(sub_list)))

val_list = set(sub_list).difference(set(train_list))
val_list = list(val_list)

train_x = seqs1h[train_list].float().flatten(1)
train_y = torch.tensor(y[train_list]).flatten()
# train_y = torch.tensor(true[train_list]).flatten()

train_x = train_x.cuda()
train_y = train_y.cuda()
train_y = train_y.flatten()

val_x = seqs1h[val_list].float().flatten(1)
val_y = torch.tensor(y[val_list]).flatten()

val_x = val_x.cuda()
val_y = val_y.cuda()
```

```python
import gpytorch

d_max = L
k_max = 8
lda_decay = 7

likelihood = gpytorch.likelihoods.GaussianLikelihood().cuda()
model = EpModel(train_x, train_y, likelihood, AA_size, L, d_max, k_max, lda_decay=lda_decay).cuda()
model.covar_module.w_kd = model.covar_module.w_kd.to(device)
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
```

```python
model.train()
likelihood.train()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

training_iter = 100

for i in range(training_iter):
    # Zero gradients from previous iteration
    optimizer.zero_grad()
    # Output from model
    output = model(train_x)
    # Calc loss and backprop gradients
    loss = -mll(output, train_y) 
    # + torch.norm(Delta2.matmul(model.covar_module.lda)[:6])
    loss.backward()
    if i % 10 == 0:
        print(f"loss = {loss.item()}")
    optimizer.step()
```

```python
lda_hat = torch.exp(model.covar_module.lda)

rho = model.covar_module.w_kd @ lda_hat
rho = rho.cpu().detach().numpy()

plt.plot(rho[:L + 1])
plt.show()

lda_hat = lda_hat.detach().cpu().numpy()

bk = np.array([binom(L, k) * (AA_size - 1)**k for k in range(1, k_max+1)])

vc = lda_hat*bk
vc = vc/vc.sum()

plt.scatter(np.arange(1, k_max + 1), vc)
plt.show()

raw_lda_opt = model.covar_module.raw_lda
torch.save(raw_lda_opt, results_path + "raw_lda.pt")
```

```python
import random 
# log_lda = torch.tensor(5*np.arange(1, k_max + 1)).float()
# raw_lda_opt = torch.nn.Parameter(log_lda.to(device))

nrep = 2
# var_list = []
# vs_list = []
# vcs_list = []
marg_efx_list = []


for pred in preds:
    for _ in range(nrep):
        sub_list = random.sample(range(len(datafile)), 20000)
        train_list = random.sample(sub_list, int(.5*len(sub_list)))
        train_x = seqs1h[train_list].float().flatten(1)
        train_y = torch.tensor(pred[train_list]).flatten()
        train_x = train_x.cuda()
        train_y = train_y.cuda()
        alpha = get_alpha(train_x, train_y, raw_lda_opt)
        total_var = get_var(train_x, raw_lda_opt, alpha)

        # vs, vcs = get_vcs(train_x, raw_lda_opt, alpha)

        marg_efx = [get_marg_efx(train_x, raw_lda_opt, alpha, k) for k in [1, 2, 3, 4]]

        # var_list.append(total_var)
        # vs_list.append(vs)
        # vcs_list.append(vcs)
        marg_efx_list.append(marg_efx)    
```

```python
marg_k = {}
for k in [1, 2, 3, 4]:
    marg_k[k] = [k * marg_efx_list[i][k - 1]/(marg_efx_list[i][k - 1].sum()) for i in range(nrep*2)]
```

### Site specific interactions - saturation curves

```python
from kernels import MargKernel, EpKernel
```

```python
import gpytorch

class MargKernel(gpytorch.kernels.Kernel):
    # the sinc kernel is stationary
    is_stationary = True

    # We will register the parameter when initializing the kernel
    def __init__(self, n_alleles, seq_length,
                 **kwargs):
        super().__init__(**kwargs)
        self.alpha = n_alleles
        self.l = seq_length
        self.d_max = d_max
        self.k_max = k_max
        self.w_d = torch.tensor([AA_size - 1, -1])

    def forward(self, x1, x2, site, **params):
        hamming_dist = (self.covar_dist(x1[:, site*AA_size:((site + 1)*AA_size)], 
                                        x2[:, site*AA_size:((site + 1)*AA_size)])**2).round()/2
        
        kernel = -1*torch.ones(*hamming_dist.shape).to(hamming_dist.device)
        kernel += AA_size * (hamming_dist == 0)
        
        return kernel
```

```python
def get_var(train_x, raw_lda, alpha):
    ker_full = EpKernel(AA_size, L, d_max, k_max, constrain_lda=False).to(device)
    ker_full.raw_lda = torch.nn.Parameter(2*raw_lda).to(device)
    ker_full.w_kd = ker_full.w_kd.to(device)
    
    K_full = ker_full(train_x, train_x).evaluate()
    K_full = K_full.detach()
    total_var = alpha.T.matmul(K_full).matmul(alpha).item()
    
    return total_var

def get_vcs(train_x, raw_lda, alpha):
    vs = []
    for k in range(k_max):
        lda_ = torch.full((k_max,), float('inf'))
        lda_[k] = 2*raw_lda[k]
        lda_ = torch.nn.Parameter(lda_)
        ker_full = EpKernel(AA_size, L, d_max, k_max, constrain_lda=False).to(device)
        ker_full.raw_lda = lda_
        K_full = ker_full(train_x, train_x).evaluate()
        K_full = K_full.detach()
        vc = alpha.T.matmul(K_full).matmul(alpha)
        vs.append(vc.item())

    vcs = np.array(vs)
    vcs = vcs/vcs.sum()
    
    return vs, vcs
```

```python
def binom(n, k):
    """Compute binomial coefficient using the log-gamma function for stability."""
    # Ensure n and k are tensors
    n = torch.as_tensor(n, dtype=torch.float)
    k = torch.as_tensor(k, dtype=torch.float)
    return torch.exp(torch.lgamma(n + 1) - torch.lgamma(k + 1) - torch.lgamma(n - k + 1))
```

```python
y = preds[0]

y = (y - y.mean())/y.std()

train_list = range(n_rand)

train_x = seqs1h[train_list].float().flatten(1)
train_y = torch.tensor(y[train_list]).flatten()

train_x = train_x.cuda()
train_y = train_y.cuda()
train_y = train_y.flatten()
```

```python
import gpytorch

d_max = L
k_max = 8
lda_decay = 6

likelihood = gpytorch.likelihoods.GaussianLikelihood().cuda()
model = EpModel(train_x, train_y, likelihood, AA_size, L, d_max, k_max, lda_decay=lda_decay).cuda()
model.covar_module.w_kd = model.covar_module.w_kd.to(device)
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
```

```python
model.train()
likelihood.train()
optimizer = torch.optim.Adam(model.parameters(), lr=0.02)

training_iter = 200

for i in range(training_iter):
    # Zero gradients from previous iteration
    optimizer.zero_grad()
    # Output from model
    output = model(train_x)
    # Calc loss and backprop gradients
    loss = -mll(output, train_y) 
    # + torch.norm(Delta2.matmul(model.covar_module.lda)[:6])
    loss.backward()
    if i % 10 == 0:
        print(f"loss = {loss.item()}")
    optimizer.step()
```

```python
lda_hat = torch.exp(model.covar_module.lda)

rho = model.covar_module.w_kd @ lda_hat
rho = rho.cpu().detach().numpy()

plt.plot(rho[:L + 1])
plt.show()

lda_hat = lda_hat.detach().cpu().numpy()

bk = np.array([binom(L, k) * (AA_size - 1)**k for k in range(1, k_max+1)])

vc = lda_hat*bk
vc = vc/vc.sum()

plt.scatter(np.arange(1, k_max + 1), vc)
plt.show()

raw_lda_opt = model.covar_module.raw_lda
torch.save(raw_lda_opt, results_path + "raw_lda.pt")
```
