```python
# site specific tokenization

def get_A2N_list(seqs):
    seqs_ = list([list(seq) for seq in seqs])
    seqs__ = pd.DataFrame(seqs_)
    alphabet_by_site = [list(seqs__.iloc[:, i].unique()) for i in range(seqs__.shape[1])]
    A2N_list = [dict(zip(alphabet, range(len(alphabet)))) for alphabet in alphabet_by_site]        
    return A2N_list

def tokenize(seq):
    numeric_sequence = [A2N_list[i][seq[i]] for i in range(len(A2N_list))]
    return numeric_sequence

def sgpu():
    GPUtil.showUtilization()

sgpu()
```

#### Define train test scheme

```python
import torch
import torch.utils.data as data

class ProtDataset(data.Dataset):
    def __init__(self, feats, labels, train=True):    
        self.train = train
        self.feats = feats
        self.labels = labels
    
    def __len__(self):
        return len(self.feats)
    
    def __getitem__(self, idx):
        
        X = self.feats[idx]
        y = self.labels[idx]
        
        return X, y
```

```python
# mutagenesis sampling

min_d = 2
sub_list = list(np.where(datafile['hd'] <= min_d)[0])
comp_list = list(np.where(datafile['hd'] > min_d)[0])

num_train = 20000
print(num_train)
num_test = 1000

# train_list = np.random.choice(sub_list, len(sub_list), replace=False)
train_list = np.random.choice(sub_list, num_train, replace=False)
test_list = np.random.choice(comp_list, num_test, replace=False)

print(len(train_list))
print(len(test_list))
```

#### Fit Linear model

```python
X = seqs1hf.float().to(device)
y = phenotypes.to(device)

X_train, y_train = X[train_list], y[train_list]
X_test, y_test = X[test_list], y[test_list]

train_dataset = ProtDataset(X_train, y_train)
train_loader = data.DataLoader(train_dataset,
                               batch_size=100,
                               shuffle=True,
                               drop_last=False)

X_train.shape
```

```python
class LinearModel(nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.fc = nn.Linear((AA_size)*L, 1)
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)                
        self.sigmoid = nn.Sigmoid()
        self.sigmoid_scale = nn.Linear(1, 1)        
        
    def forward(self, x):
        x = self.fc(x)
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scale(x)        
            
        return x
```

```python
x, y = next(iter(train_loader))
model = LinearModel().to(device)
output = model(x)

plt.hist(output.detach().cpu().numpy())
plt.show()
```

```python
import torch.optim as optim
import torch
import torch.nn as nn


from scipy.stats import pearsonr
learning_rate = 0.001
epochs = 100

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

for epoch in range(epochs):
    total_loss = 0
    for batch_inputs, batch_targets in train_loader:
        model.train()
        optimizer.zero_grad()
        outputs = model(batch_inputs)
        loss = criterion(outputs, batch_targets)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(train_loader)}")
        model.eval()
        pred, true = model(X_test.flatten(1)).flatten().detach().cpu().numpy(), y_test.flatten().detach().cpu().numpy()
        print(pearsonr(pred, true)[0]**2)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np


import torch
import torch.nn as nn

class MultiHeadAttentionLayer(nn.Module):
    def __init__(self, hidden_dim, num_heads, dropout):
        super(MultiHeadAttentionLayer, self).__init__()
        self.multihead_attn = nn.MultiheadAttention(hidden_dim, num_heads, dropout)
        # self.multihead_attn = MultiheadAttention(hidden_dim, hidden_dim, num_heads)        
        self.layer_norm = nn.LayerNorm(hidden_dim)
        
    def forward(self, x):
        attn_output, _ = self.multihead_attn(x, x, x)
        x = x + attn_output
        # x = self.layer_norm(x + attn_output)
        return x

class CustomTransformer(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        
        self.transformer_layers = nn.ModuleList([
            MultiHeadAttentionLayer(hidden_dim, num_heads, dropout) for _ in range(num_layers)
        ])
        # self.fc = nn.Linear(hidden_dim, 1)
        self.fc = nn.Linear(hidden_dim*L, 1)        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)

    def forward(self, x):

        x = self.embedding(x)
        x = x.permute(1, 0, 2)  # seq_len x batch x hidden_dim
        
        for layer in self.transformer_layers:
            x = layer(x)
            
        # x = torch.mean(x, dim=0)  # batch x hidden_dim
        x = x.permute(1, 0, 2)
        x = x.flatten(1)
        # print(x.shape)
        x = self.fc(x)  # batch x 1 (scalar)
        x = self.sigmoid_norm(x)
#         mean = torch.mean(x)
#         std = torch.std(x)
#         x = (x - mean) / std
        
        # x = self.phi_scaling(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x

sequence_length = L
input_dim = AA_size*L
output_dim = 1
hidden_dim = 128
num_layers = 1
num_heads = 4
dropout = 0.3

model = CustomTransformer(input_dim, hidden_dim, num_layers, num_heads, dropout).to(device)
```

#### Fit MLP

```python
X = seqs1hf.float().to(device)
y = phenotypes.to(device)

X_train, y_train = X[train_list], y[train_list]
X_test, y_test = X[test_list], y[test_list]

train_dataset = ProtDataset(X_train, y_train)
train_loader = data.DataLoader(train_dataset,
                               batch_size=1000,
                               shuffle=True,
                               drop_last=False)

X_train.shape
```

```python
mlp_width = X.shape[-1]
```

```python
class MLP(nn.Module):
    def __init__(self, n_layers, hidden_dim):
        super(MLP, self).__init__()
        self.n_layers = n_layers
        self.hidden_dim = hidden_dim
        self.fc_in = nn.Linear(mlp_width, self.hidden_dim)
        self.fc_out = nn.Linear(hidden_dim, 1)    
        self.fc_layers = nn.ModuleList([nn.Linear(hidden_dim, hidden_dim) for _ in range(self.n_layers)])
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)                
        self.sigmoid = nn.Sigmoid()
        self.sigmoid_scale = nn.Linear(1, 1)   
        self.m = nn.Tanh()
        
    def forward(self, x):
        x = self.fc_in(x)
        x = self.m(x)
        for l in self.fc_layers:
            x = l(x)
            x = self.m(x)
        x = self.fc_out(x)
        # x = self.m(x)
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        # x = self.sigmoid_scale(x)        
            
        return x
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc = nn.Linear(mlp_width, hidden_dim)
        self.fc1 = nn.Linear(hidden_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim)        
        self.fc3 = nn.Linear(hidden_dim, 1)        
        self.fc_norm = nn.BatchNorm1d(hidden_dim, affine=False)                        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)                
        self.sigmoid = nn.Sigmoid()
        self.sigmoid_scale = nn.Linear(1, 1)   
        self.m = nn.Tanh()
        
    def forward(self, x):
        x = self.fc(x)
        x = self.m(x)
        x = self.fc_norm(x)        
        x = self.fc1(x)
        x = self.m(x)
        x = self.fc_norm(x)        
        x = self.fc2(x)
        x = self.m(x)
        x = self.fc_norm(x)        
        x = self.fc3(x)

        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        # x = self.sigmoid_scale(x)        
            
        return x
```
