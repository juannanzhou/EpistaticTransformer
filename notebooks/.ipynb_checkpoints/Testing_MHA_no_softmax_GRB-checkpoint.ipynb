```python
# site specific tokenization

def get_A2N_list(seqs):
    seqs_ = list([list(seq) for seq in seqs])
    seqs__ = pd.DataFrame(seqs_)
    alphabet_by_site = [list(seqs__.iloc[:, i].unique()) for i in range(seqs__.shape[1])]
    A2N_list = [dict(zip(alphabet, range(len(alphabet)))) for alphabet in alphabet_by_site]        
    return A2N_list

def tokenize(seq):
    numeric_sequence = [A2N_list[i][seq[i]] for i in range(len(A2N_list))]
    return numeric_sequence

def sgpu():
    GPUtil.showUtilization()

sgpu()
```

```python
class CustomMultiHeadAttention(nn.Module):
    def __init__(self, embed_dim=512, n_heads=8, dropout=.3):
        """
        Args:
            embed_dim: dimension of embeding vector output
            n_heads: number of self attention heads
        """
        super(CustomMultiHeadAttention, self).__init__()

        self.embed_dim = embed_dim    #512 dim
        self.n_heads = n_heads   #8
        self.single_head_dim = int(self.embed_dim / self.n_heads)   #512/8 = 64  . each key,query, value will be of 64d
       
        #key,query and value matrixes    #64 x 64   
        self.query_matrix = nn.Linear(self.single_head_dim , self.single_head_dim ,bias=False)  # single key matrix for all 8 keys #512x512
        self.key_matrix = nn.Linear(self.single_head_dim  , self.single_head_dim, bias=False)
        self.value_matrix = nn.Linear(self.single_head_dim ,self.single_head_dim , bias=False)
        self.out = nn.Linear(self.n_heads*self.single_head_dim ,self.embed_dim) 

    def forward(self,key,query,value,mask=None):    #batch_size x sequence_length x embedding_dim    # 32 x 10 x 512
        
        """
        Args:
           key : key vector
           query : query vector
           value : value vector
           mask: mask for decoder
        
        Returns:
           output vector from multihead attention
        """
        batch_size = key.size(0)
        seq_length = key.size(1)
        
        # query dimension can change in decoder during inference. 
        # so we cant take general seq_length
        seq_length_query = query.size(1)
        
        # 32x10x512
        key = key.view(batch_size, seq_length, self.n_heads, self.single_head_dim)  
        #batch_size x sequence_length x n_heads x single_head_dim = (32x10x8x64)
        query = query.view(batch_size, seq_length_query, self.n_heads, self.single_head_dim) #(32x10x8x64)
        value = value.view(batch_size, seq_length, self.n_heads, self.single_head_dim) #(32x10x8x64)
       
        k = self.key_matrix(key)       # (32x10x8x64)
        q = self.query_matrix(query)   
        v = self.value_matrix(value)

        q = q.transpose(1,2)  # (batch_size, n_heads, seq_len, single_head_dim)    # (32 x 8 x 10 x 64)
        k = k.transpose(1,2)  # (batch_size, n_heads, seq_len, single_head_dim)
        v = v.transpose(1,2)  # (batch_size, n_heads, seq_len, single_head_dim)
       
        # computes attention
        # adjust key for matrix multiplication
        k_adjusted = k.transpose(-1,-2)  #(batch_size, n_heads, single_head_dim, seq_ken)  #(32 x 8 x 64 x 10)
        product = torch.matmul(q, k_adjusted)  #(32 x 8 x 10 x 64) x (32 x 8 x 64 x 10) = #(32x8x10x10)
      
        
        # fill those positions of product matrix as (-1e20) where mask positions are 0
        if mask is not None:
             product = product.masked_fill(mask == 0, float("-1e20"))

        #divising by square root of key dimension
        product = product / math.sqrt(self.single_head_dim) # / sqrt(64)

        #applying softmax
        scores = product
        # scores = F.softmax(product, dim=-1)
 
        #mutiply with value matrix
        scores = torch.matmul(scores, v) 
        
        #concatenated output
        concat = scores.transpose(1,2).contiguous().view(batch_size, seq_length_query, self.single_head_dim*self.n_heads)
        
        output = self.out(concat) 
       
        return output

class CustomTransformer(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        self.dropout1 = nn.Dropout(dropout)
        self.transformer_layers = nn.ModuleList([
            CustomMultiHeadAttention(hidden_dim, num_heads, dropout) for _ in range(num_layers)
        ])
        self.fc = nn.Linear(hidden_dim*L, 1)        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)

    def forward(self, x):

        x = self.embedding(x)
        # x = x.permute(1, 0, 2)  # seq_len x batch x hidden_dim
        
        for layer in self.transformer_layers:
            x = layer(x, x, x)
            x = self.dropout(x)
            
        # x = x.permute(1, 0, 2)
        x = x.flatten(1)
        x = self.fc(x)  
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import torch
import torch.nn as nn

class MultiHeadAttentionLayer(nn.Module):
    def __init__(self, hidden_dim, num_heads, dropout):
        super(MultiHeadAttentionLayer, self).__init__()
        self.multihead_attn = nn.MultiheadAttention(hidden_dim, num_heads, dropout)
        # self.multihead_attn = MultiheadAttention(hidden_dim, hidden_dim, num_heads)        
        self.layer_norm = nn.LayerNorm(hidden_dim)
        
    def forward(self, x):
        attn_output, _ = self.multihead_attn(x, x, x)
        x = x + attn_output
        # x = self.layer_norm(x + attn_output)
        return x

class CustomTransformer(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        
        self.transformer_layers = nn.ModuleList([
            MultiHeadAttentionLayer(hidden_dim, num_heads, dropout) for _ in range(num_layers)
        ])
        # self.fc = nn.Linear(hidden_dim, 1)
        self.fc = nn.Linear(hidden_dim*L, 1)        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)

    def forward(self, x):

        x = self.embedding(x)
        x = x.permute(1, 0, 2)  # seq_len x batch x hidden_dim
        
        for layer in self.transformer_layers:
            x = layer(x)
            
        # x = torch.mean(x, dim=0)  # batch x hidden_dim
        x = x.permute(1, 0, 2)
        x = x.flatten(1)
        # print(x.shape)
        x = self.fc(x)  # batch x 1 (scalar)
        x = self.sigmoid_norm(x)
#         mean = torch.mean(x)
#         std = torch.std(x)
#         x = (x - mean) / std
        
        # x = self.phi_scaling(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x
```

```python
class CustomTransformer_original(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer_original, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        
        self.transformer_layers = nn.ModuleList([
            CustomMultiHeadAttention(hidden_dim, num_heads, dropout) for _ in range(num_layers)
        ])
        self.fc = nn.Linear(hidden_dim*L, 1)
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):

        x = self.embedding(x)
        
        for layer in self.transformer_layers:
            x = layer(x, x, x)
            x = self.dropout(x)
            
        x = x.flatten(1)
        x = self.fc(x)  
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x
```

```python
class CustomTransformer(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        
        self.transformer_layer0 = CustomMultiHeadAttention(hidden_dim, num_heads, dropout) 
        self.transformer_layer1 = CustomMultiHeadAttention(hidden_dim, num_heads, dropout) 
        self.fc = nn.Linear(hidden_dim*L, 1)        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)
        self.dropout = nn.Dropout(dropout)
        self.embeds_t = nn.Linear(hidden_dim, hidden_dim)


    def forward(self, inputs):

        embeds = self.embedding(inputs)
        x = embeds
        embeds = self.embeds_t(embeds)        
        x = self.transformer_layer0(x, x, x)
        x = self.dropout(x)
        x = self.transformer_layer1(x, x, embeds)
        x = self.dropout(x)
            
        x = x.flatten(1)
        x = self.fc(x)  
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x
```

```python
class CustomTransformer_1_layer(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_heads, dropout):
        super(CustomTransformer_1_layer, self).__init__()

        self.embedding = nn.Embedding(input_dim, hidden_dim)
        
        self.transformer_layer0 = CustomMultiHeadAttention(hidden_dim, num_heads, dropout) 
        self.transformer_layer1 = CustomMultiHeadAttention(hidden_dim, num_heads, dropout) 
        self.fc = nn.Linear(hidden_dim*L, 1)        
        self.sigmoid_norm = nn.BatchNorm1d(1, affine=False)        
        self.sigmoid = nn.Sigmoid()
        self.phi_scaling = nn.Linear(1, 1)
        self.sigmoid_scaling = nn.Linear(1, 1)
        self.dropout = nn.Dropout(dropout)
        self.embeds_t = nn.Linear(hidden_dim, hidden_dim)


    def forward(self, inputs):

        embeds = self.embedding(inputs)
        x = embeds
        x = self.transformer_layer0(x, x, x)
        x = self.dropout(x)
            
        x = x.flatten(1)
        x = self.fc(x)  
        x = self.sigmoid_norm(x)
        x = self.sigmoid(x)
        x = self.sigmoid_scaling(x)

        return x
```

```python
import optuna
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from scipy.stats import pearsonr
learning_rate = 0.001
epochs = 200
num_heads = 4
num_layers = 4
sequence_length = L
input_dim = AA_size*L
output_dim = 1
dropout_init = 0.3
batch_size_init = 1000
hidden_dim_h_init = 32

def objective(trial):
    # Define the hyperparameters to be optimized
    # hidden_dim_h = trial.suggest_int('hidden_dim_h', hidden_dim_h_init, 10, 50)
    # dropout = trial.suggest_float('dropout', dropout_init, 0.05, 0.35)
    # batch_size = trial.suggest_int('batch_size', batch_size_init, 100, 1200)

    hidden_dim_h = trial.suggest_int('hidden_dim_h', 10, 200)
    dropout = trial.suggest_float('dropout', 0.05, 0.35)
    batch_size = trial.suggest_int('batch_size', 100, 1200)
    
    model = CustomTransformer(input_dim, hidden_dim_h*num_heads, num_layers, num_heads, dropout).to(device)
    
    train_loader = data.DataLoader(train_dataset,
                                   batch_size=batch_size,
                                   shuffle=True,
                                   drop_last=False)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    r2_test = []
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for batch_inputs, batch_targets in train_loader:
            optimizer.zero_grad()
            outputs = model(batch_inputs)
            loss = criterion(outputs, batch_targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        if epoch % 10 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Loss: {total_loss/len(train_loader)}")
            model.eval()
            pred, true = model(X_test.flatten(1)).flatten().detach().cpu().numpy(), y_test.flatten().detach().cpu().numpy()
            print(pearsonr(pred, true)[0]**2)
            r2_test.append(pearsonr(pred, true)[0]**2)
            
    return np.array(r2_test).max()
```

```python
# Create and run the Optuna study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Print the best hyperparameters
best_trial = study.best_trial
print("Best Trial:")
print(f"  Value: {best_trial.value:.4f}")
print("  Params: ")
for key, value in best_trial.params.items():
    print(f"    {key}: {value}")

# You can then use the best hyperparameters to train your final model.
```
