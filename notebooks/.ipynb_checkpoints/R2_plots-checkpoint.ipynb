```python
result_path = "../results/output/"
```

```python
dirs = os.listdir(result_path)
```

```python
r2_all = {}

for d in dirs:
    r2s = pd.read_csv(result_path + d + "/R2s.csv", index_col = 0).to_numpy().T[0]
    r2s = np.pad(r2s, (0, 4 - len(r2s)), mode='constant')
    r2_all[d] = r2s
    
```

```python
R2_df = pd.DataFrame(r2_all).T

R2_df.columns = ["Linear", "Pairwise", "4th-Order", "8th-Order"]

R2_df = R2_df.sort_index()
```

```python
R2_df
```

```python
data_list = pd.read_csv("/blue/juannanzhou/ProteinLLE/Data/Data_prepared/Protein_set.txt", header=None)
data_list = [file.split(".")[0] for file in data_list.iloc[:, 0]]
```

```python
data_summary = pd.read_csv("../results/Dataset summary-Table 1.csv")
```

```python
data_summary
```

```python
name_dict = {}
for i in range(len(data_summary)):
    name_dict[data_summary.iloc[i]['OldName']] = data_summary.iloc[i]['Name'].split("_")[1]
```

```python
data_summary
```

```python
R2_ = {}

for data in data_list:
    R2_[data] = R2_df[[data in ID for ID in R2_df.index]]

point_size = 50
plt.rcParams.update({'font.size': 20})

plt.rcParams['font.family'] =  'Nimbus Sans'
```

```python
fig, axs = plt.subplots(3, 4, figsize=(20, 15),
                       gridspec_kw={'hspace': 0.5, 'wspace': 0.5})  # 2 rows, 1 column

for i in range(len(data_list)): 
    
    data_name = data_list[i]

    plotdat = R2_[data_name]

    plotdat.index = ["20%", "50%", "80%"]


    # axs[int(i/4), i%4].plot(plotdat, 'o', linestyle='-', markersize=10)
    for column in plotdat.columns:
        axs[int(i/4), i%4].plot(plotdat.index, plotdat[column], label=column, marker='o', linestyle='-', markersize=10)
        # plt.plot(plotdat.index, plotdat[column], label=column)

    axs[int(i/4), i%4].set_title(name_dict[data_name])
    axs[int(i/4), i%4].set_ylim(0.5, 1)
    axs[int(i/4), i%4].set_ylabel('$R^2$')
    if i == 0:
        axs[int(i/4), i%4].legend(fontsize='smaller')
        axs[int(i/4), i%4].legend(loc='center left', bbox_to_anchor=(-.8, .8), fontsize=14)

plt.show()    
```
